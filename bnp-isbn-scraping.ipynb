```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm.notebook import tqdm
import time
import random
import re
import xml.etree.ElementTree as ET
```

```python
# Lista de URL para recolha de dados

#http://urn.bn.pt/isbn/mods/xml?id=9789896417529

static = 'http://urn.bn.pt/isbn/mods/xml?id='

isbn_list = []
# open file and read the content in a list
with open('isbn-list.txt', 'r') as filehandle:
    isbn_list = [isbn.rstrip() for isbn in filehandle.readlines()]

urllist = [static+str(isbn) for isbn in isbn_list]
```

```python
#urllist_test = urllist[23:25]

maindata = pd.DataFrame()

for i in tqdm(urllist):
    headers = requests.utils.default_headers()
    headers.update({'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/63.0'})
    URL = i

    page = requests.get(URL)
    
    soup = BeautifulSoup(page.content, "lxml")

    book = {}
    try:
        book["ISBN"] = soup.identifier.string
        book["titulo"] = soup.title.string
        book["autor"] = soup.namepart.string
    #book["id_autor"] = soup.namepart
        book["editora"] = soup.publisher.string
        book["data_publicacao"] = soup.dateissued.string
        book["url_BNP"] = i
    except:
        print("failed for", i, "\n")
        file = open(r"not-working.txt","a")
        file.write( i + "\n")
        file.close()

    book_data = pd.DataFrame(book, index=[0])
    
    maindata = pd.concat([maindata, book_data])
    time.sleep(random.randint(1,2))
    
print("Já está")
maindata
```
